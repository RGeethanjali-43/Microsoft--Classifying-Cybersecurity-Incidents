**LOGISTIC REGRESSION**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, classification_report, confusion_matrix

# Load your data
# final = pd.read_csv('your_data.csv')  # Uncomment and modify this line if you're loading data from a file

# Sample n rows from the dataset
n = 100000  # Specify the number of samples you want to test with
sampled_data = final.sample(n=n, random_state=42)

# Separate features and target variable
X = sampled_data.drop(columns=['IncidentGrade'])  # Replace 'target_column' with your actual target column name
y = sampled_data['IncidentGrade']  # The target column

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and fit the logistic regression model
logreg = LogisticRegression(max_iter=1000)  # Increase max_iter if needed
logreg.fit(X_train, y_train)

# Predict on the test set
y_pred = logreg.predict(X_test)

# Calculate the F1 score
f1 = f1_score(y_test, y_pred, average='weighted')  # Use 'macro' or 'weighted' depending on the scenario

# Classification matrix
class_matrix = classification_report(y_test, y_pred)



# Print results
print(f"F1 Score: {f1}")
print("Classification Report:\n", class_matrix)

```

**CROSS VALIDATION**

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import make_scorer, f1_score

# Load your data
# final = pd.read_csv('your_data.csv')  # Uncomment and modify this line if you're loading data from a file

# Sample n rows from the dataset
n = 100000  # Specify the number of samples you want to test with
sampled_data = final.sample(n=n, random_state=42)

# Separate features and target variable
X = sampled_data.drop(columns=['IncidentGrade'])  # Replace 'IncidentGrade' with your actual target column name
y = sampled_data['IncidentGrade']  # The target column

# Initialize the Decision Tree model
dtree = DecisionTreeClassifier(random_state=42)

# Define the F1 score as the scoring metric
f1_scorer = make_scorer(f1_score, average='weighted')

# Perform cross-validation
cv_scores = cross_val_score(dtree, X, y, cv=5, scoring=f1_scorer)

# Print results
print(f"Cross-Validation F1 Scores: {cv_scores}")
print(f"Mean Cross-Validation F1 Score: {cv_scores.mean()}")
print(f"Standard Deviation of Cross-Validation F1 Scores: {cv_scores.std()}")
```

**ROC CURVE**

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt

# Assuming you've already trained your model and have predictions
# Predict probabilities for each class
y_pred_proba = xgb.predict_proba(X_test)

# Binarize the output for multi-class ROC curve
n_classes = len(np.unique(y_test))
y_binarized = label_binarize(y_test, classes=np.unique(y_test))

# Plot ROC curve
plt.figure(figsize=(5, 5))

for i in range(n_classes):
    fpr, tpr, _ = roc_curve(y_binarized[:, i], y_pred_proba[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'ROC curve for class {i} (area = {roc_auc:.2f})')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Multi-Class Classification')
plt.legend()
plt.show()
```
